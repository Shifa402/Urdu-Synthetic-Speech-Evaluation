```python
import pandas as pd

df = pd.read_excel("UrduSER_Description.xlsx", header=0)

# assign correct Urdu headers manually
df.columns = [
    "تشریح",
    "جذبات",
    "صنف",
    "اداکار",
    "اداکار_آئی_ڈی",
    "نمونہ_کی_شکل",
    "نمونہ_کی_شرح",
    "فارمیٹ",
    "دورانیہ_سیکنڈز",
    "فائل_کا_نام",
    "سیریل_نمبر"
]
print(df.head())
```

```python
print(df.head())
```

```python
df = df.rename(columns={
    "تشریح": "text",
    "جذبات": "emotion",
    "صنف": "gender",
    "اداکار": "actor",
    "اداکار_آئی_ڈی": "actor_id",
    "نمونہ_کی_شکل": "bit_depth",
    "نمونہ_کی_شرح": "sample_rate",
    "فارمیٹ": "format",
    "دورانیہ_سیکنڈز": "duration_sec",
    "فائل_کا_نام": "file_name",
    "سیریل_نمبر": "sr_no"
})
```

```python
df.info()
df.head()
df["emotion"].value_counts()
```

```python
ranges = {
    "غصہ": [1,2,3,4,5,6,7,8,9,19],
    "خوف": [63,64,65,66,67,68,69,70,71,72],
    "بوریت": [90,91,92,93,94,95,96,97,98,99],
    "ناگواری": [81,82,83,84,85,86,87,88,89,90],
    "خوشی": [64,65,66,67,68,69,70,71,72,73],
    "افسردہ": [54,55,56,57,60,61,62,63,64,65]
}
```

```python
# subtract 1 from every index
ranges = {
    emotion: [i - 1 for i in indices]
    for emotion, indices in ranges.items()
}

print(ranges)
```

```python
result = []

for emotion, indices in ranges.items():
    group = df[df["emotion"] == emotion]

    # Select specific rows by position (not index label)
    selected = group.iloc[indices]

    result.append(selected)

final_df = pd.concat(result).reset_index(drop=True)

print(final_df[["emotion", "sr_no", "file_name", "text"]])
```

```python
import google.generativeai as genai
import io, wave
import numpy as np
import soundfile as sf
import time

# =========================
# Setup
# =========================
genai.configure(api_key="")
model = genai.GenerativeModel("gemini-2.5-flash-preview-tts")

scores = []
results = []

# =========================
# Use your dataframe
# =========================
dataset = final_df.reset_index(drop=True)

for i in range(8,10):

    text = dataset.loc[i, "text"]
    file_name = dataset.loc[i, "file_name"]

    gen_path = f"gen_{file_name}.wav"

    # =========================
    # Gemini TTS
    # =========================
    response = model.generate_content(
        f"Please read this text aloud in Urdu clearly: {text}",
        generation_config={"response_modalities": ["AUDIO"]}
    )

    try:
        audio_bytes = response.candidates[0].content.parts[0].inline_data.data

        byte_io = io.BytesIO()
        with wave.open(byte_io, 'wb') as wav_file:
            wav_file.setnchannels(1)
            wav_file.setsampwidth(2)
            wav_file.setframerate(24000)
            wav_file.writeframes(audio_bytes)

        byte_io.seek(0)

        with open(gen_path, "wb") as f:
            f.write(byte_io.read())

    except Exception as e:
        print(f"Error on sample {file_name}: {e}")
        continue

    print(f"Saved: gen_{file_name}.wav | {text[:40]}...")

    # =========================
    # Rate limiting
    # =========================
    if (i + 1) % 3 == 0:
        print("Waiting for 72 seconds...")
        time.sleep(72)
```
