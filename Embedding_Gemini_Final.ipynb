# Embedding - Book

```python
# =========================
# 1. Install Dependencies
# =========================
!pip install resemblyzer torchaudio scipy numpy

# =========================
# 2. Imports
# =========================
import os
import numpy as np
from scipy.spatial.distance import cosine
from resemblyzer import VoiceEncoder, preprocess_wav

# ========================= # # 3. Upload ZIP File # # ========================= #
from google.colab import files
uploaded = files.upload()
zip_path = list(uploaded.keys())[0]
extract_path = "/content/data"
# # ========================= # # 4. Extract ZIP # # =========================
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
  zip_ref.extractall(extract_path) # print("Extracted!")

# =========================
# 3. Paths
# =========================
ref_dir = None
gen_dir = None

for root, dirs, files in os.walk(extract_path):
    for d in dirs:
        if d.lower() == "reference":
            ref_dir = os.path.join(root, d)
        elif d.lower() == "generated_gemini_tts":
            gen_dir = os.path.join(root, d)

print("Reference folder:", ref_dir)
print("Generated folder:", gen_dir)

if ref_dir is None or gen_dir is None:
    raise Exception("Could not find 'reference' or 'generated' folders!")

# =========================
# 4. Load Resemblyzer Encoder
# =========================
encoder = VoiceEncoder()

# =========================
# 5. Embedding Function
# =========================
def get_embedding(file_path):
    wav = preprocess_wav(file_path)  # handles resampling + mono conversion
    embedding = encoder.embed_utterance(wav)
    return embedding

# =========================
# 6. Compute Similarity
# =========================
results = []

ref_files = sorted([f for f in os.listdir(ref_dir) if f.endswith(".wav")])

for file_name in ref_files:
    ref_path = os.path.join(ref_dir, file_name)
    gen_path = os.path.join(gen_dir, file_name)

    if not os.path.exists(gen_path):
        print(f"Skipping {file_name}, not found in generated/")
        continue

    ref_emb = get_embedding(ref_path)
    gen_emb = get_embedding(gen_path)

    similarity = 1 - cosine(ref_emb, gen_emb)

    results.append((file_name, similarity))
    print(f"{file_name}: {similarity:.4f}")

# =========================
# 7. Average Score
# =========================
avg_similarity = np.mean([r[1] for r in results])
print("\nAverage Cosine Similarity:", avg_similarity)
```

```python
# =========================
# 10. Save & Download CSV
# =========================
import pandas as pd
from google.colab import files

# Create dataframe
df = pd.DataFrame(results, columns=["file_name", "cosine_similarity"])

# Add average row (optional)
df.loc[len(df)] = ["AVERAGE", avg_similarity]

# Save CSV
csv_path = "/content/similarity_results_book_gemini.csv"
df.to_csv(csv_path, index=False)

print("CSV saved at:", csv_path)

# Download
files.download(csv_path)
```
